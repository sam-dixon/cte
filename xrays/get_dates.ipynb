```python
import os
from astropy.io import fits
```

```python
dirname = 'camera2'
for root, dirs, files in os.walk(dirname):
    for fname in files:
        if 'fits' not in fname.split('.'):
            continue
        path = os.path.join(root, fname)
        header = fits.getheader(path)
        print(path, header['DATE'])
```
